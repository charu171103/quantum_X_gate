# X Gate

```python
from qiskit import QuantumCircuit
from qiskit.quantum_info import Statevector
from qiskit.visualization import plot_state_qsphere, plot_bloch_multivector
import matplotlib.pyplot as plt
# Ensure plots are displayed inline
%matplotlib inline
```

```python
def initialize_state(circuit, state):
    if state == '|0>':
        pass  # Already initialized to |0> by default
    elif state == '|1>':
        circuit.x(0)  # X gate to flip to |1>
    elif state == '|+>':
        circuit.h(0)  # Hadamard to create |+>
    elif state == '|->':
        circuit.h(0)
        circuit.z(0)  # Z gate to create |->
    elif state == '|i>':
        circuit.h(0)
        circuit.s(0)  # S gate to create |i>
    elif state == '|-i>':
        circuit.h(0)
        circuit.sdg(0)  # S-dagger gate to create |-i>
        
def simulate_x_gate(input_state):
    # Create a single qubit quantum circuit
    qc = QuantumCircuit(1)

    # Initialize the input state
    initialize_state(qc, input_state)

    # Apply the X gate
    qc.x(0)

    # Get the statevector
    statevector = Statevector.from_instruction(qc)

    # Return the circuit and statevector
    return qc, statevector
# Test simulate_x_gate function for |0> state
qc, statevector = simulate_x_gate('|0>')

# Print results
print("Quantum Circuit:")
print(qc.draw())

print("\nStatevector:")
print(statevector.data)

# Simulate the input state |0>
qc, statevector = simulate_x_gate('|0>')

# Visualize input and output
print("Input State: |0>")
print("Output Statevector:", statevector.data)

# Visualize Q-Sphere
print("Q-Sphere Visualization (Output):")
plot_state_qsphere(statevector)
plt.show()

# Visualize Bloch Sphere
print("Bloch Sphere Visualization (Output):")
plot_bloch_multivector(statevector)
plt.show()

# Show circuit
print("Quantum Circuit:")
print(qc.draw())
```

```python
# Simulate the input state |-> 
qc, statevector = simulate_x_gate('|->')
print("Input State: |->")
print("Output Statevector:", statevector.data)

# Visualize Q-Sphere
print("Q-Sphere Visualization (Output):")
fig1 = plot_state_qsphere(statevector)
plt.show(fig1)

# Visualize Bloch Sphere
print("Bloch Sphere Visualization (Output):")
fig2 = plot_bloch_multivector(statevector)
plt.show(fig2)

# Show circuit
print("Quantum Circuit:")
print(qc.draw())
```
